```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.metrics import pairwise_distances,euclidean_distances
from sklearn.manifold import MDS
from sklearn.decomposition import PCA
import pandas as pd
from scipy.spatial.distance import pdist,squareform,euclidean
from keras.datasets import cifar10
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train.reshape(50000,3072)
```

```python
X_train = X_train.reshape(50000,3072)
X_test = X_test.reshape(10000,3072)
```

```python
data_set = np.vstack((X_train,X_test))
label_set = np.vstack((y_train,y_test))
```

```python
df = pd.DataFrame(np.hstack((data_set,label_set)),columns = [(i for i in range(3072))].append(3072))
```

```python
mean_images = df.groupby(3072).mean().values
```

```python
data_dict = {}
for i in range(10):
    key = "data" + str(i)
    data_dict[key] = df[df[3072] == i].drop(columns = [3072]).values
```

```python
pca_dict = {}
project_dict = {}
error_dict = {}
for key, value in data_dict.items():
    model = PCA(20).fit(value)
    pca_dict[key] = model.components_
    project_dict[key] = model.inverse_transform(model.transform(value))
    error_dict[key] = (euclidean(project_dict[key].flatten(),value.flatten())**2) / len(value)
```

```python
plt.rcParams["axes.grid"] = False
plt.figure(figsize=(15,12))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(mean_images[i].reshape((32,32,3)) / (255.0))
plt.savefig('mean.png')
plt.show()
plt.close()
```

```python
class_name = {
    0: 'airplane',
    1: 'automobile',
    2: 'bird',
    3: 'cat',
    4: 'deer',
    5: 'dog',
    6: 'frog',
    7: 'horse',
    8: 'ship',
    9: 'truck',
}
df_error = pd.DataFrame(list(error_dict.values()),index = class_name.values())
```

```python
plt.figure(figsize=(3,4))
plt.subplot(2,1,2)
df_error.plot.bar(color = 'gbrwy')
plt.xlabel("Category")
plt.ylabel("mean error")
plt.title("bar plot for PCA error of each category")
plt.legend().remove()
plt.tight_layout()
plt.savefig('error.png',dpi = 120)
plt.show()
plt.close()
```

```python
D = pairwise_distances(mean_images)
D = D ** 2
```

```python
np.savetxt("partb_distances.csv", D, delimiter=",")
```

```python
A = np.identity(10) - (np.ones(10).reshape((10,1)) @ (np.ones(10).reshape(1,10)))/10
W = (A@D@A.T) / -2
s,v = np.linalg.eig(W)
result = v[:,:2] @ np.diag(s[:2] ** 0.5)
plt.figure()
for i in range(10):
    plt.scatter(result[i, 0], result[i, 1])
    plt.text(result[i, 0] - 100,result[i, 1] -100,class_name[i])
plt.axis('equal');
plt.xlabel("First component")
plt.ylabel("Second component")
plt.title("plot for PCoA dimension reduction")
plt.savefig("PartB.png")
plt.show()
plt.close()
```

```python
def calculate_difference(a,b):
    return (euclidean(a.flatten(),b.flatten()) ** 2 / len(a))
error_diff_array = []
for i in range(10):
    keyi = "data" + str(i)
    for j in range(10):
        keyj = "data" + str(j)
        a = data_dict[keyi]
        value = data_dict[keyj]
        modeli = PCA(20).fit(value)
        modelj = PCA(20).fit(a)
        modeli.mean_ = mean_images[i]
        modelj.mean_ = mean_images[j]
        projecti = modeli.inverse_transform(modeli.transform(a))
        projectj = modelj.inverse_transform(modelj.transform(value))
        error_diff_array.append((calculate_difference(projecti,a) + calculate_difference(projectj,value)) / 2)
error_matrix = (np.array(error_diff_array).reshape(10, 10))  # resha
error_matrix
```

```python
np.savetxt("partc_distances.csv", error_matrix, delimiter=",")
```

```python
A = np.identity(10) - (np.ones(10).reshape((10,1)) @ (np.ones(10).reshape(1,10)))/10
W = (A@error_matrix@A.T) / -2
s,v = np.linalg.eig(W)
result = v[:,:2] @ np.diag(s[:2] ** 0.5)
plt.figure()
for i in range(10):
    plt.scatter(result[i, 0], result[i, 1])
    plt.text(result[i, 0] - 50 ,result[i, 1] - 50,class_name[i])
plt.axis('equal');
plt.xlabel("First component")
plt.ylabel("Second component")
plt.title("plot for dimension reduction with our special distance matrix")
plt.savefig("PartC.png")
plt.show()
plt.close()
```
